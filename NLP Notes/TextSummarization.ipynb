```python
# Step-2 calculate wordFrequency
## Commenting STOPWORDS AND PUNCTUATION CONDITION AS FOR THE PRESENTATION SLIDE EXAMPLE
for word in textDoc:
    if word.text.lower() not in stopwords and word.text not in punctuation:
      if word.text not in wordFrequency.keys():
          wordFrequency[word.text]=1
      else:
          wordFrequency[word.text]=wordFrequency[word.text]+1

wordFrequency
```

```python
# maxFrequency=max(wordFrequency.values())

maxFrequency= len(wordFrequency.keys())

maxFrequency
```

#### Step-3 Update @word_frequency as representing .....Normalization of word frequency....

```python
# Step-3 Update @word_frequency as representing .....Normalization of word frequency....
for word in wordFrequency.keys():
    wordFrequency[word]=wordFrequency[word]/maxFrequency

print(wordFrequency)
```

```python
# Step-4 Sentence Tokenization
sentences=[sent for sent in textDoc.sents]
sentences
```

####  STEP 5: Calculate sentence scores for each sentence

```python
## STEP 5: Calculate sentence scores for each sentence
sentenceScores={}
```

```python
type(sentences)
```

```python
type(wordFrequency)
```

```python
#### Computing the sentence score
for sentence in sentences:
    print("For ", sentence)
    if sentence not in sentenceScores.keys():
        sentenceScores[sentence]=0
    for word in sentence:
        if word.text in wordFrequency.keys():
            print(word, wordFrequency[word.text])
            sentenceScores[sentence] +=wordFrequency[word.text]
    print("The score is",sentenceScores[sentence])
print(sentenceScores)
```

```python
percentageOfSentences= 0.7
summaryLength=int(len(sentences)* percentageOfSentences)
summaryLength
```

#### STEP 6: Rank the sentences using @sentence_scores

```python
# STEP 6: Rank the sentences using @sentence_scores
summary =nlargest(summaryLength,sentenceScores,key= sentenceScores.get)
summary
```
